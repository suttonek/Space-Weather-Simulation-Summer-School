# Investigation with 1-year SYM-H Data

In the following code, we read the SYM-H index from the file of "omni_min2013.asc" which contains all the derived parameters (solar winds, kp, AE indices, SYM-H, etc.) Then the distribution of day when SYM-H< -100 nT is made. Note that the position of the SYM-H index is different from that in "omni_min_case_2013.lst" file.

```python
import read_ascii_file as read
import numpy as np
import matplotlib.pyplot as plt

file = '../data/omni_min2013.asc'
index = -5
time,year,day,hour,minute,symh = read.read_ascii_file(file,index)
fig,ax = plt.subplots()

lp = symh <-100
ax.hist(day[lp],bins = np.arange(1,366,2))
ax.set_xlabel('Day of year in 2013')
ax.set_ylabel('Events')
```

Print the list of time for these events.

```python
print(time[lp])
```

We can do more with datetime. For example, print the dates of these events.

```python
import datetime as dt
print(time[0].date())
```

```python
date_1 = []
for iT in time[lp]:
    date_1.append(iT.date())
print('-->>> Date of these events:')
print(np.unique(np.asarray(date_1)))
```
