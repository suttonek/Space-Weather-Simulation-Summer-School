# Call User-defined Functions from Other Files

The following code shows how to call the read_ascii_file function in read_ascii_file.py

```python
import read_ascii_file as read  # import module
import matplotlib.pyplot as plt

file = '../data/omni_min_case_2013.lst'
index = 4 

# call read_ascii_file function in read_ascii_file.py
time,symh,year,day,hour,minute = read.read_ascii_file(file,index) 
fig,ax = plt.subplots()

ax.plot(time,symh,marker='.',c='gray',
        label='All Events',alpha=0.5)
lp = symh <-100
ax.plot(time[lp],symh[lp],marker='+',
        linestyle='',
        c= 'tab:orange',
        label='<-100 nT',
        alpha=0.6)
ax.set_xlabel('Year of 2013')
ax.set_ylabel('SYMH (nT)')
ax.grid(True)
ax.legend()
```
